### 1.2 Calculating descriptors
Dictionary with keys for entry ID and molecular descriptos was created with empty lists as values.

Mobley IDs and the following molecular descriptors were calcualted and appneded to the dictionary.
1. Molecular weight (MW)
1. Crippen atomic permeability partition coefficient (AlogP)
1. Number of hydrogen bond acceptors (HBA)
1. Number of hydrogen bond donors (HBD)
1. __Active Groups__

```python
import glob
from rdkit.Chem import SDMolSupplier, Descriptors, Crippen, Lipinski, Fragments

# Empty dictionary for ligand IDs and corresponding descriptros.
data = {'ID': [], 'MW (Da)': [], 'AlogP': [], 
        'HBA': [], 'HBD': [], '#NH_OH': [], '#Active groups': []}


def num_active_groups(mol):
    '''Counts the number of active groups in a molecule according to 
    biologically active functional groups.'''
    
    counter = 0
    
    counter += int(Fragments.fr_COO(mol))
    # Carbonyl excluding carboxylic acids
    counter += int(Fragments.fr_C_O_noCOO(mol))
    # Primary amines
    counter += int(Fragments.fr_NH2(mol))
    # Secondary amines
    counter += int(Fragments.fr_NH1(mol))
    # Thiols
    counter += int(Fragments.fr_SH(mol))
    # Sulfonamides
    counter += int(Fragments.fr_sulfonamd(mol))
    # Aromatic rings
    counter += int(Lipinski.NumAromaticRings(mol))
    # Aliphatic hydroxy groups
    counter += int(Fragments.fr_Al_OH(mol))
    # Aromatic hydroxy groups
    counter += int(Fragments.fr_Ar_OH(mol))
        
    return counter

# Location of SDF files.
sdf_dr = './sdffiles/'

for sdf in glob.glob(sdf_dr + '*.sdf'):
    
    # add ligand Mobley IDs
    ID = sdf.strip('jupyter_notebook/sdffiles/.')
    data['ID'].append(ID)
    
    # add calculated descriptors
    suppl = SDMolSupplier(sdf)
    for mol in suppl:
        data['MW (Da)'].append(Descriptors.MolWt(mol))
        data['AlogP'].append(Crippen.MolLogP(mol))
        data['HBA'].append(Lipinski.NumHAcceptors(mol))
        data['HBD'].append(Lipinski.NumHDonors(mol))
        data['#NH_OH'].append(Lipinski.NHOHCount(mol))
        data['#Active groups'].append(num_active_groups(mol))
```

Use Pandas to store calculated descriptors and for simple Matplotlib implimentation.

### 1.4 Descriptor histograms
A function was written to plot a histogram for any specified DataFrame column.

```python
from rdkit.Chem import Draw
from rdkit.Chem import AllChem

def plot_hist(property):
    '''Enter column name from df as property 
    and specify range() as bin range.'''
    
    min_x = int(df[property].min())
    max_x = int(df[property].max())
    if property == 'MW (Da)':
        bin_range = range(0, 501, 50)
    else:
        bin_range = range(min_x, max_x)
    df[property].plot(kind='hist', bins=bin_range, color='white', 
                      edgecolor='black', xticks=bin_range)
    plt.xlabel(property)
    plt.show()
```

### 1.5 Display structures with the minimum and maximum values for descriptors calculated

```python
def get_min_max(dataframe, descriptor_lst):
    'Retrieve the minimum and maximum values and for a all descriptor.'
    
    min_max_dr = []
    legends = []
    
    for desc in descriptor_lst:
        # Sort main df according to descriptor.
        sorted_df = dataframe.sort_values(desc)
        sorted_df = sorted_df.reset_index(drop=True)
        
        # Get ID and descriptor df index.
        ID_index = df.columns.get_loc('ID')
        descriptor_index = df.columns.get_loc(desc)
        
        # Provide ID for first and last sturcutre and descriptor.
        ID = [sorted_df.iloc[0, ID_index], sorted_df.iloc[-1, ID_index]]
        for i in range(2):
            min_max_dr.append(sdf_dr + ID[i] + '.sdf')    

        # Provide value for first and last sturcutre and descriptor.
        value = [round(sorted_df.iloc[0, descriptor_index], 2), 
                 round(sorted_df.iloc[-1, descriptor_index], 2)]
        legends.append('Lowest ' + str(desc) + ' = ' + str(value[0]))
        legends.append('Highest ' + str(desc) + ' = ' + str(value[1]))
        
    return min_max_dr, legends


def draw_structure_panel(sdf_suppl, legend):
    'Draw RDKit.Draw in panel format.'

    suppl = [SDMolSupplier(sdf) for sdf in sdf_suppl]
    mols = [x[0] for x in suppl if x is not None]
    for mol in mols: tmp = AllChem.Compute2DCoords(mol)
    return Draw.MolsToGridImage(mols, molsPerRow=2, subImgSize=(200, 200), legends=legend)  


min_max_dr, legend =  get_min_max(df, descriptors)
draw_structure_panel(min_max_dr, legend)
```

### 2.2 Determining discrepancy

```python
diff = len(df3) - len(df2)
print('Discrepancy between SAMPL4_Gurthire in FreeSolv and SI is:', diff)
```

Visualise missing ligands represented by a dash '-'.

```python
def check_MW(df2, df3):
    '''Iterate down MW columns in two dataframes and check if values at the same index
    are equal. If not, a new row of zeros is inserted at the infringing index.
    
    Note: at this point, df3 is the longer dataframe of the two.'''
    
    for i, MW1, MW2 in zip(range(len(df3)), df2.loc[:, 'MW'], df3.loc[:, 'MW']):
        if MW1 != MW2 and MW1 != '-':
            start_upper = 0
            end_upper = i
            start_lower = i
            end_lower = df2.shape[0]
            upper_half = [*range(start_upper, end_upper, 1)]
            lower_half = [*range(start_lower, end_lower, 1)]
            lower_half = [x.__add__(1) for x in lower_half]
            index_ = upper_half + lower_half
            df2.index = index_
            df2.loc[i] = '-'
            df2 = df2.sort_index()

            break
    
    return df2


for i in range(len(df3)):
    df2 = check_MW(df2, df3)

df4 = [[MW1, MW2] for MW1, MW2 in zip(df2.loc[:, 'MW'], df3.loc[:, 'MW'])]
df4 = pd.DataFrame(df4, columns=['FreeSolv', 'Guthrie SI'])
df4
```

### 3.4 Offsets
Create dataframe containing absolute hydration energy offsets between experimental and calculated values within FreeSolv.

```python
# Training set data: SAMPL4_Guthrie entries (2.1.2) removed from FreeSolv (2.1.1).
train_data = [df1.iloc[i] for i in range(len(df1))  if df1.loc[i, exp_ref_col] != ref]

# New training set DataFrame not containing SAMPL4_Guthrie entires.
df8 = pd.DataFrame(train_data)
df8
```

Calculate absolute hydration free energies, with associated linear propogation uncertainties.

```python
# Column names
ID = df8.loc[:, 'compound id (and file prefix)']
exp_val = df8.loc[:, 'experimental value (kcal/mol)']
exp_err = df8.loc[:, 'experimental uncertainty (kcal/mol)']
calc_val = df8.loc[:, 'Mobley group calculated value (GAFF) (kcal/mol)']
calc_err = df8.loc[:, 'calculated uncertainty (kcal/mol)']

# New nested list containing IDs and offsets
offsets = []
for name, exp, err1, calc, err2 in zip(ID, exp_val, exp_err, calc_val, calc_err):
    offset = exp - calc
    offset = np.absolute(offset)
    error = (err1**2 + err2**2)**0.5
    offsets.append([name, offset, round(error, 2)])

df9 = pd.DataFrame(offsets, columns=['ID', 'dGoffset (kcal/mol)', 'uncertainty (kcal/mol)'])
df9
```

Discovered non-numeric cells in train_df during:

\>\>\> train_df.apply(pd.to_numeric)

This was due to mordred descriptor calculation error except messages. A function was written to check for non-numeric cells and drop the respective column.

```python
def check_dataframe_is_numeric(dataframe):
    'Iterate over all columns and check if numeric.'

    columns_dropped = 0

    for col in dataframe.columns:
        for x in dataframe.loc[:, col]:
            try:
                float(x)
            except ValueError:
                # Uncomment below to view infringing column.
                # print('\'{}\' in {} is not a float.'.format(x, col))
                columns_dropped += 1
                dataframe = dataframe.drop(columns=col)
                break

    print('Number of columns dropped:', columns_dropped)
    return dataframe

train_df = check_dataframe_is_numeric(train_df)
train_df
```
