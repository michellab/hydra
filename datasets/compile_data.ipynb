# Compile data
Data to be compiled:
1. pFP
1. dFEAT
1. Offsets

```python
import pandas as pd

# Data file paths
pFP_path = '../pFP/dFP_output/dFP_train.csv'
dFEAT_path = '../dFEAT/dFeatures_output/dFEAT_train.csv'
offsets_path = './dGhydr_offsets.csv'

df_pFP = pd.read_csv(pFP_path)
df_dFEAT = pd.read_csv(dFEAT_path)
df_offsets = pd.read_csv(offsets_path)

df_master = df_pFP
df_master = df_master.join(df_dFEAT.set_index('ID'), on='ID')
df_master = df_master.join(df_offsets.set_index('ID'), on='ID')
df_master
```

```python
import os

if not os.path.exists('./train_compiled'):
    os.makedirs('./train_compiled')

save_loc = './train_compiled/dGhydr_train.csv'
df_master.to_csv(path_or_buf=save_loc, index=False)

print('Completed writing all dGhydr compiled training set to CSV.')
```
